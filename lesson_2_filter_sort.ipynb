```python
import pandas as pd
```

```python
df = pd.read_excel("data/world_population_excel_workbook.xlsx")
```

```python
df
```

```python
df[df['Rank'] <= 10]
```

```python
visegrad_group = ["Poland", "Czech Republic", "Slovakia", "Hungary"]
df[df['Country'].isin(visegrad_group)]
```

```python
df[df['Country'].str.contains('United')]
```

```python
df[df['Country'].str.contains('[P-Z].*land$')]  # you can use regex!
```

```python
df2 = df.set_index('Country')
df2 = df.set_index(['Country'])  # I think those two do the same
```

```python
df2
```

```python
# The three below do the exact same thing
df2.filter(['CCA3', 'Continent'])
df2.filter(['CCA3', 'Continent'], axis=1)  # axis 1 is default
df2.filter(items=['CCA3', 'Continent'])

df2.filter(items=['Country', 'CCA3', 'Continent'])  # you can add the index column also, 'Country' in this case
```

```python
# Filtering rows. This only works because we set 'Country' as index
df2.filter(items=['Zimbabwe'], axis=0)
```

```python
df2.filter(items=visegrad_group, axis=0)
```

```python
df2.filter(like="United", axis=0)
```

```python
df2.filter(regex="[P-Z].*land$", axis=0)  # regex parameter instead of like parameter!
```

```python
df2.loc['United States']
```

```python
df2.loc[['United States']]
```

```python
df2.loc[visegrad_group]
```
