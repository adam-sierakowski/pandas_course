```python
import pandas as pd
```

```python
df = pd.read_csv("data/Flavors.csv")
df
```

```python
# Four ways to fix a typo
df.at[8, "Flavor"] = "Chocolate Fudge Brownie"
df.iat[8, 0] = "Chocolate Fudge Brownie"
df.loc[8, "Flavor"] = "Chocolate Fudge Brownie"
df.iloc[8, 0] = "Chocolate Fudge Brownie"
```

```python
df
```

```python
group_by_frame = df.groupby("Base Flavor")
group_by_frame
```

```python
group_by_frame.mean(numeric_only = True)
```

```python
df.mean(numeric_only = True)
```

```python
group_by_frame.count()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').sum()
```

```python
df.groupby('Base Flavor').sum(numeric_only=True)
```

```python
df.groupby('Base Flavor').agg(
    {
        "Flavor Rating": ["mean", "max", "count", "sum"],
        "Texture Rating": ["mean", "max", "count", "sum"]
    }
)
```

```python
df.groupby('Base Flavor').agg(
    {
        "Flavor": lambda x: "|".join(x),
    }
)
```

```python
df.groupby('Base Flavor').agg(
    {
        "Flavor": "sum"
    }
)
```

```python
df.groupby('Base Flavor').agg(
    {
        "Flavor": list
    }
)
```

```python
df.groupby('Base Flavor').agg(
    {
        "Flavor": list,
        "Liked": list,
        "Flavor Rating": "mean",
        "Texture Rating": "mean",
        "Total Rating": "mean"
    }
)
```

```python
df.groupby(['Base Flavor', 'Liked']).mean(numeric_only=True)
```

```python
df.groupby(['Base Flavor', 'Liked']).agg(
    {
        "Flavor": list,
        "Liked": "first",  # might as well comment this out, but here's the .first method!
        "Flavor Rating": "mean",
        "Texture Rating": "mean",
        "Total Rating": "mean"
    }
)
```

```python
df.groupby(['Base Flavor', 'Liked']).agg(
    {
        "Flavor": list,
        "Liked": lambda x: list(x)[2],  # the third element
        "Flavor Rating": "mean",
        "Texture Rating": "mean",
        "Total Rating": "mean"
    }
)
```
