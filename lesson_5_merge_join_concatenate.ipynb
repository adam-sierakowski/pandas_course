```python
import pandas as pd
```

```python
df1 = pd.read_csv("data/LOTR.csv")
```

```python
df2 = pd.read_csv("data/LOTR 2.csv")
```

```python
df1
```

```python
df2
```

## .merge()

```python
df1.merge(df2, how="inner")
df1.merge(df2, how="inner", on=["FellowshipID", "FirstName"])
# df1.merge(df2)  # inner join is the default
```

```python
df1.merge(df2, how="inner", on="FellowshipID")
```

```python
df1.merge(df2, how="outer")
```

```python
df1.merge(df2, how="left")
```

```python
df1.merge(df2, how="right")
```

```python
df1.merge(df2, how="cross")
# compare each value from the left to all of the values on the right
# not very useful, except on the interviews
```

## .join()

```python
df1.join(df2)  # won't work as it doesn't have default arguments as .merge does
```

```python
df1.join(df2, on="FellowshipID", how="outer", lsuffix="_left", rsuffix="_right")
# usually you don't use it like that, but with indexes
```

```python
df4 = df1.set_index("FellowshipID").join(
    df2.set_index("FellowshipID"),
    lsuffix="_left",
    rsuffix="_right",
    how="outer"
)
df4
```

```python
df1.merge(df2, how="outer").set_index("FellowshipID")
```

## .concatenate()

```python
df5 = pd.concat([df1, df2])
df5
```

```python
df5.iloc[4]
```

```python
pd.concat([df1, df2], join="inner")
```

```python
pd.concat([df1, df2], join="outer", axis=1)
```
