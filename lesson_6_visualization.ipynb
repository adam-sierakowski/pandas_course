```python
import pandas as pd
```

```python
import matplotlib as plt  # works without importing, but can't set style
import matplotlib.style
```

```python
print(plt.style.available)
plt.style.use('bmh')
```

```python
df = pd.read_csv("data/Ice Cream Ratings.csv")
df = df.set_index("Date")
```

```python
df
```

```python
df.plot(kind="barh", xlabel="Rating", title="Ice cream ratings", figsize=(10,5))
```

```python
df.plot(kind="line")  # "line" is the default
```

```python
df.plot(kind="barh", subplots=True, title="Ice cream ratings")
```

```python
df.reset_index().plot(kind="scatter", x="Date", y="Overall Rating")
```

```python
df.plot(kind="bar", stacked=True)
```

```python
df["Overall Rating"].plot(kind="bar")
```

### You can also use the different kinds as methods

```python
df.plot.barh()
```

```python
df.reset_index().plot.scatter(x="Date", y="Overall Rating", s=200, c="red")
```

```python
df.boxplot()
```

```python
df.describe()
```

```python
df.plot.area(figsize=(10,5))
```

```python
df.plot.pie(y="Overall Rating")
```

```python
df.plot.pie(subplots=True)
```
